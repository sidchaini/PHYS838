```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

sns.set_theme(context="paper", style="whitegrid", font_scale=1.1,
              rc = {'xtick.bottom': True, 'xtick.top': True, 'ytick.left': True, 'ytick.right': True,
                   'axes.grid': True,  'axes.edgecolor': '.15'})
```

```python
df = pd.read_csv("data_100k.csv")
```

```python
df['rarad'] = np.radians(df['ra'])
df['decrad'] = np.radians(df['dec'])

figure = plt.figure()
# figure.patch.set_facecolor('black') # setting plot background to dark color
ax = figure.add_subplot(111, projection='mollweide')
plt.scatter(df['rarad'], df['decrad'], s=1, color='red')
ax.set_xlabel('Right Ascension')
ax.set_ylabel('Declination')
plt.show()
```

```python
df['rarad'] = np.radians(df['ra'])
df['decrad'] = np.radians(df['dec'])

figure = plt.figure()
# figure.patch.set_facecolor('black') # setting plot background to dark color
ax = figure.add_subplot(111, projection='mollweide')
plt.scatter(df['rarad'], df['decrad'], c=df['redshift'], cmap='viridis', s=1)
ax.set_xlabel('Right Ascension')
ax.set_ylabel('Declination')
plt.colorbar(label='Redshift')
plt.show()
```
