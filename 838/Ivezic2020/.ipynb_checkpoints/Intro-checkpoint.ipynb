#### Figure 1.3

```python
from astroML.datasets import fetch_sdss_specgals
data = fetch_sdss_specgals()
data.shape
```

```python
data.dtype.names[:5]
```

```python
data = data[:10000] # truncate data
u = data['modelMag_u']
r = data['modelMag_r']
rPetro = data['petroMag_r']
ax = plt.axes ()
ax.scatter(u-r, rPetro , s=4, lw=0, c='k')
ax.set_xlim (1, 4.5)
ax.set_ylim (18.1 , 13.5)
ax.set_xlabel('$u - r$')
ax.set_ylabel('$r_{petrosian}$')
```

#### Figure 1.4

```python
from astroML.datasets import fetch_dr7_quasar
data = fetch_dr7_quasar()
data.shape
```

```python
data.dtype.names[:5]
```

```python
data = data [:10000]
r = data['mag_r']
i = data['mag_i']
z = data['redshift']
ax = plt.axes ()
ax.scatter(z, r - i, s=4, c='black', linewidth =0)
ax.set_xlim (0, 5)
ax.set_ylim (-0.5, 1.0)
ax.set_xlabel('redshift')
ax.set_ylabel('r-i')
```

#### Figure 1.5

```python
from astroML.datasets import fetch_sdss_sspp
data = fetch_sdss_sspp()
data.shape
```

```python
data.dtype.names[:5]
```

```python
data = data [:10000]
rpsf = data['rpsf'] # make some reasonable cuts
data = data [( rpsf > 15) & (rpsf < 19)]
logg = data['logg']
Teff = data['Teff']
```

```python
ax = plt.axes()
ax.scatter(Teff , logg , s=4, lw=0, c='k')
ax.set_xlim(8000 , 4500)
ax.set_ylim(5.1, 1)
ax.set_xlabel(r'$\mathrm{T_{eff}\ (K)}$')
ax.set_ylabel(r'$\mathrm{log_{10}[g / (cm/s^2)]}$')
```

#### Figure 1.6

```python
from astroML.datasets import fetch_sdss_S82standards
data = fetch_sdss_S82standards ()
data.shape
```

```python
data.dtype.names[:5]
```

```python
data = data [:10000]
g = data['mmu_g'] # g-band mean magnitude
r = data['mmu_r'] # r-band mean magnitude
i = data['mmu_i'] # i-band mean magnitude
```

```python
ax = plt.axes ()
ax.scatter(g - r, r - i, s=4, c='black', linewidth =0)
ax.set_xlabel('g - r')
ax.set_ylabel('r - i')
```

#### Figure 1.8

```python
from astroML.datasets import fetch_moving_objects
data = fetch_moving_objects(Parker2008_cuts =True)
data.shape
```

```python
data.dtype.names[:5]
```

```python
data = data[:10000]
a = data['aprime']
sini = data['sin_iprime']
```

```python
ax = plt.axes ()
ax.scatter(a, sini , s=4, c='black', linewidth =0)
ax.set_xlabel('Semi -major Axis (AU)')
ax.set_ylabel('Sine of Inclination Angle')
```
